```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras import models, datasets

# Load CIFAR-10 dataset
(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
x_train_cifar = x_train_cifar.astype('float32') / 255
x_test_cifar = x_test_cifar.astype('float32') / 255

# Load the model from Experiment 3
model_path_exp3 = "/Users/pc/Documents/Exp3/fine_tuned_model.keras"
model_exp3 = models.load_model(model_path_exp3)

# Load the fine-tuned model
fine_tuned_model_path = "/Users/pc/Documents/Exp3/fine_tuned_model.keras"
fine_tuned_model = models.load_model(fine_tuned_model_path)

# Evaluate the model from Experiment 3 on CIFAR-10 test set
test_loss_exp3, test_accuracy_exp3 = model_exp3.evaluate(x_test_cifar, y_test_cifar, verbose=0)
print("Experiment 3 - CIFAR-10 Test Accuracy:", test_accuracy_exp3)

# Evaluate the fine-tuned model on CIFAR-10 test set
test_loss_fine_tuned, test_accuracy_fine_tuned = fine_tuned_model.evaluate(x_test_cifar, y_test_cifar, verbose=0)
print("Fine-Tuned Model - CIFAR-10 Test Accuracy:", test_accuracy_fine_tuned)

# Save the results to a CSV file
results_df = pd.DataFrame({
    "Model": ["Experiment 3", "Fine-Tuned Model"],
    "Test Accuracy": [test_accuracy_exp3, test_accuracy_fine_tuned]
})
results_df.to_csv("experiment_results.csv", index=False)


```

```python
import matplotlib.pyplot as plt

# Test accuracies
accuracies = [test_accuracy_exp3, test_accuracy_fine_tuned]
models = ['Experiment 3', 'Fine-Tuned Model']

# Plotting the point plot
plt.figure(figsize=(8, 6))
plt.plot(models, accuracies, marker='o', linestyle='-', color='blue')
plt.xlabel('Model')
plt.ylabel('Test Accuracy')
plt.title('Comparison of Test Accuracies')
plt.ylim(0, 1)  # Limit y-axis to between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
